Compare the reliability of the following schemes, assuming an exponential failure occurence with following values:
*   MTTFA = 900 h
*   MTTFB = 7000 h
*   MMTFC = 1000 h








```python
import numpy as np
import plotly.graph_objects as go
t = np.arange(0, 5001, 1)
```

```python
def PloatFigure(R1_t,R2_t,titleName):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=t, y=R1_t, mode='lines', name='Sys1', line=dict(color='red')))
    fig.add_trace(go.Scatter(x=t, y=R2_t, mode='lines', name='Sys2', line=dict(color='cyan')))
    fig.update_layout(title=titleName,
                      xaxis_title='Time (h)',
                      yaxis_title='Reliability',
                      legend=dict(x=1, y=1, traceorder='normal'))
    fig.show()
```

# **Configurazione Parametri**

```python
if __name__ == "__main__":
  #Calcolo Lambda in funzione degli MTTF
  lambda_A = 1 / 500
  lambda_B = 1 / 9000
  lambda_C = 1 / 10000
```

Per ogni Rx definico il corisspettivo rx_t in funzione del tempo $ e^{-\lambda t} $:


```python
ra_t = np.exp(-lambda_A * t)
rb_t = np.exp(-lambda_B * t)
rc_t = np.exp(-lambda_C * t)
```

# **Confronto prima coppia di sistemi**

1.JPG

Questi sistemi possono essere esspressi mediante le seguenti equazioni:

```python
R1_t = 1 - (1 - ra_t * rb_t) * (1 - ra_t * rc_t)
R2_t = ra_t * (1 - (1 - rb_t) * (1 - rc_t))
```

```python
PloatFigure(R1_t,R2_t,'Confronto prima coppia di sistemi')
```

In rosso abbiamo il sistema 1, notiamo quindi che il primo sistema ha una
reliability migliore.

# **Confronto seconda coppia di sistemi**

2.JPG

Questi sistemi possono essere esspressi mediante le seguenti equazioni:

```python
R1_t = ra_t*(1-(1-ra_t)*(1-rb_t))
R2_t = ra_t
```

```python
PloatFigure(R1_t,R2_t,'Confronto seconda coppia di sistemi')
```

Dal grafico notiamo che il sistema 2  ovviamente migliore: questa cosa ce
la aspettavamo per la propriet che dice che la reliability totale di un sistema  minore o uguale della reliability dei singoli componenti in serie

# **Confronto terza coppia di sistemi**

3.JPG

Questi sistemi possono essere esspressi mediante le seguenti equazioni:

```python
R1_t = ra_t*rb_t*(1-(1-ra_t)*(1-rb_t))
R2_t = ra_t*rb_t
```

```python
PloatFigure(R1_t,R2_t,'Confronto terza coppia di sistemi')
```

Il discorso  lo stesso del caso precedente: il sistema a destra composto dalla serie di due componenti, quello a sinistra dalla serie di due componenti pi un parallelo; il secondo sistema  il migliore.

# **Confronto quarta coppia di sistemi**



Questi sistemi possono essere esspressi mediante le seguenti equazioni:

```python
R1_t = 1 - (1 - ra_t) * (1 - ra_t * rb_t)
R2_t = ra_t
```

```python
PloatFigure(R1_t,R2_t,'Confronto quarta coppia di sistemi')
```

Dal grafico notiamo che il sistema 1 ha una reliability migliore: questa cosa
ce la aspettavamo per la propriet che dice che il sistema in parallelo avr una reliability maggiore o uguale del componente con maggiore reliability.
