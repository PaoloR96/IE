<h1> PCA e Clustering </h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

def inter_devianza_cluster(grouped_df, work_df):            
    inter = 0
    for group_name, df_group in grouped_df:
        for series_name, series in df_group.items():
            baricentro = work_df[series_name].mean()
            centroide  = series.mean()        
            val = np.power(centroide-baricentro, 2)
            inter +=  (val*series.size)
    return inter
    
def intra_devianza_cluster(grouped_df):
    intra_df = grouped_df.aggregate(lambda x: intra_devianza_val(x))
    vals = intra_df.sum()
    return vals.sum()
    
def intra_devianza_val(points):
    arr = np.array(points, dtype=float)      
    centr = arr.mean()
    arr -= centr
    arr = np.power(arr, 2)
    return np.sum(arr)

def clustering_df(work_df, min_clustering, max_clustering, c_link, pca_usate, var_pca):  
    tempDf = pd.DataFrame(columns=['Intra','Inter','PcaUsate','NumCluster', 'VarPersa%', 'Var PCA%', 'VarConservata%'])
    for num_cluster in range(min_clustering, max_clustering, 10):
        fl = fcluster(c_link, num_cluster, criterion='maxclust')
        grouped_df = work_df.groupby(fl)
        intra = intra_devianza_cluster(grouped_df)
        inter = inter_devianza_cluster(grouped_df, work_df)
        var_persa = (intra)/(intra+inter)
        tempDf.loc[len(tempDf)] = [intra, inter, pca_usate, num_cluster, 100*var_persa, var_pca, (1-var_persa)*var_pca*100]
    return tempDf

def get_work_df(df, num_col, i):
    return df.iloc[:,num_col:num_col+i]

def read_file(path_file, num_col, var_pca,  min_num_pca = 3, max_num_pca=7, min_clustering = 20, max_clustering=100, debug = False):
    df = pd.read_csv(path_file, decimal=",")
    res_df = pd.DataFrame(columns=['Intra','Inter','PcaUsate', 'NumCluster', 'VarPersa%', 'Var PCA%', 'VarConservata%'])
    for i in range(min_num_pca, max_num_pca+1):
        work_df = get_work_df(df, num_col, i)
        if debug == True:print(work_df)
        c_link = linkage(work_df,  metric='euclidean', method='ward')
        tempDf = clustering_df(work_df, min_clustering, max_clustering, c_link, i, var_pca[i-1])
        res_df = pd.concat([res_df,tempDf]) 
    return res_df
```

```python
def draw_graph(df):
    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
    i = 0
    min_pca = df['PcaUsate'].min()
    for group_name, df_group in df.groupby('PcaUsate'):
        Num_cluster_arr = df_group['NumCluster'].values.tolist()
        Var_finale_arr =  df_group['VarConservata%'].values.tolist()
        label = 'Pca usate - ' + str(i + min_pca)
        plt.plot(Num_cluster_arr, Var_finale_arr,"o", color=colors[i], label=label) 
        i += 1
    plt.xlabel("Num Cluster") 
    plt.ylabel("Var Finale")
    plt.grid()
    plt.legend() 
    plt.show() 
    
def generate_random_workload(path_file, pca_usate, num_cluster, num_col):
    df = pd.read_csv(path_file, dtype= np.float64, decimal=",")
    work_df = get_work_df(df, num_col, pca_usate)
    df = df.iloc[:,num_col:num_col+pca_usate]
    c_link = linkage(work_df,  metric='euclidean', method='ward')
    fl = fcluster(c_link, num_cluster, criterion='maxclust') 
    return df.groupby(fl).sample(n=1).reset_index()

def generate_most_workload(path_file, pca_usate, num_cluster, num_col):
    df = pd.read_csv(path_file, decimal=",")
    work_df = get_work_df(df, num_col, pca_usate)
    df = df.iloc[:,num_col:num_col+pca_usate]
    c_link = linkage(work_df,  metric='euclidean', method='ward')
    fl = fcluster(c_link, num_cluster, criterion='maxclust') 
    gb = df.groupby(fl)['label']
    final_df = gb.apply(top_value_count).reset_index().rename(columns=dict(level_0 = 'cluster', level_1 = 'richiesta', level_2='label', label='count_label'))
    return final_df

def top_value_count(x, n=3):
    return x.value_counts().head(n)
```
