```python
import numpy as np 
import pylab 
import scipy.stats as stats
import matplotlib.pyplot as plt
import csv
import pingouin as pg
```

```python
import pandas as pd

def DF_subs(path_file_1, path_file_2):
    df_1 = pd.read_csv(path_file_1, dtype= np.float64, decimal=",")
    df_2 = pd.read_csv(path_file_2, dtype= np.float64, decimal=",")
    df3 = pd.DataFrame(columns=df_1.columns.values.tolist())
    for series_name, series in df_1.items():
        df3[series_name] = df_1[series_name] - df_2[series_name]
    return df3

def qq_plot(series_name, series):
        plt.title(series_name)
        ax = pg.qqplot(series, dist='norm', confidence=.95)
        test = stats.shapiro(series)
        plt.text(0.2, 0.1, test)
        plt.show()

def do_test(df_1, df_2):
    for series_1, series_2 in zip(df_1.items(), df_2.items()):
        test1 = stats.shapiro(series_1[1])
        test2 = stats.shapiro(series_2[1])
        qq_plot(series_1[0], series_1[1])
        qq_plot(series_2[0], series_2[1])
        print(stats.ranksums(series_1[1], series_2[1]))
        print('A varianza diversa: ',stats.ttest_ind(series_1[1], series_2[1], equal_var = False))
        print('A varianza uguale: ',stats.ttest_ind(series_1[1], series_2[1], equal_var = True))
```
